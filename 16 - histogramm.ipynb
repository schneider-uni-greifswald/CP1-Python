#Histogramme


```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
from scipy.stats import norm
%matplotlib inline
```

```python
data=np.genfromtxt('data/histogramm.txt')  #Hier der Pfad zu Euren Daten eintragen
```

```python
print(data[:5])
```

```python
plt.hist(data)
plt.xlabel('x / unit')
plt.ylabel('y / unit')
```

```python
plt.hist(data, bins=30) #Zahl der bins auf 30 setzen
plt.xlabel('x / unit')
plt.ylabel('y / unit')
```

```python
plt.hist(data, bins=30, density=True) #Normierung, Fl√§che des Histogramms=1
plt.xlabel('x / unit')
plt.ylabel('y / unit')
```

```python
#Kummulative Verteiling
plt.hist(data, bins=30, density=True, cumulative=True)
plt.xlabel('x / unit')
plt.ylabel('y / unit')
```

```python
#Mittelwert, Standardabweichung der Verteilung
mean=np.mean(data)
std_dev=np.std(data)
print(mean)
```

```python
n, bins, patches=plt.hist(data, bins=30, density=True)

#Gausskurve in das Histogramm zeichnen
gauss = norm.pdf(bins, mean, std_dev)
plt.plot(bins, gauss, 'k', linewidth=2)
```
